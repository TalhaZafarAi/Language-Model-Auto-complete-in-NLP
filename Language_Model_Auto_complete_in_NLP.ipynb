# **Tokenize Data**

```python
def tokenize_data(data):
    sentences = data.split('\n')
    tokenized_sentences = [nltk.word_tokenize(sentence.lower()) for sentence in sentences]
    return tokenized_sentences

# Example usage:
tokenized_data = tokenize_data(data)
print(tokenized_data[:5])  # Print the first 5 tokenized sentences to check
```

```python
from sklearn.model_selection import train_test_split

def train_test_split_data(tokenized_data, test_size=0.2):
    train_data, test_data = train_test_split(tokenized_data, test_size=test_size, random_state=42)
    return train_data, test_data

# Example usage:
train_data, test_data = train_test_split_data(tokenized_data)
print(f"Train size: {len(train_data)}, Test size: {len(test_data)}")
print('\n\n', test_data)
```

# **Words with N frequency**

```python
def get_closed_vocabulary(word_counts, threshold):
    closed_vocab = {word for word, count in word_counts.items() if count >= threshold}
    return closed_vocab

# Example usage:
threshold = 5
closed_vocab = get_closed_vocabulary(word_counts, threshold)
print(f"Closed vocabulary size: {len(closed_vocab)}")
```

# **Replace OOV with UNK**

```python
def replace_oov_with_unk(tokenized_data, closed_vocab, unk_token='<unk>'):
    replaced_data = [[word if word in closed_vocab else unk_token for word in sentence] for sentence in tokenized_data]
    return replaced_data

# Example usage:
replaced_train_data = replace_oov_with_unk(train_data, closed_vocab)
replaced_test_data = replace_oov_with_unk(test_data, closed_vocab)
print(replaced_train_data[:5])  # Print the first 5 replaced tokenized sentences to check
```

# **Preprocess Data**

```python
def preprocess_data(data, threshold, test_size=0.2):
    sentences = split_to_sentences(data)
    tokenized_sentences = tokenize_sentences(sentences)
    train_data, test_data = train_test_split_data(tokenized_sentences, test_size)
    word_counts = count_words(train_data)
    closed_vocab = get_closed_vocabulary(word_counts, threshold)
    train_data_replaced = replace_oov_with_unk(train_data, closed_vocab)
    test_data_replaced = replace_oov_with_unk(test_data, closed_vocab)
    return train_data_replaced, test_data_replaced, closed_vocab

# Example usage:
threshold = 5
train_data_processed, test_data_processed, vocab = preprocess_data(data, threshold)
print(f"Processed train size: {len(train_data_processed)}, Processed test size: {len(test_data_processed)}")
```

# **Count N Grams**

```python
from collections import defaultdict

def count_ngrams(data, n):
    ngrams = defaultdict(int)
    for sentence in data:
        sentence = ['<s>'] * (n-1) + sentence + ['<e>']
        for i in range(len(sentence) - n + 1):
            ngram = tuple(sentence[i:i+n])
            ngrams[ngram] += 1
    return ngrams

# Example usage:
n = 2
bigrams = count_ngrams(train_data_processed, n)
print(list(bigrams.items())[:10])  # Print the first 10 bigrams to check
```

# **Estimate Probability of Single Word**

```python
def estimate_probability(word, previous_ngram, ngram_counts, nplus1gram_counts, vocab_size, k=1):
    previous_ngram = tuple(previous_ngram)
    previous_ngram_count = ngram_counts.get(previous_ngram, 0)
    nplus1gram = previous_ngram + (word,)
    nplus1gram_count = nplus1gram_counts.get(nplus1gram, 0)
    probability = (nplus1gram_count + k) / (previous_ngram_count + k * vocab_size)
    return probability

# Example usage:
vocab_size = len(vocab) + 1  # Include <unk>
previous_ngram = ['i', 'eat']
word = 'scrambled'
probability = estimate_probability(word, previous_ngram, bigrams, count_ngrams(train_data_processed, 3), vocab_size)
print(f"Probability of '{word}' given '{previous_ngram}': {probability}")
```

# **Estimate Probability of all Words**

```python
def estimate_probabilities(previous_ngram, ngram_counts, nplus1gram_counts, vocab, k=1):
    vocab_size = len(vocab) + 1  # Include <unk>
    probabilities = {}
    for word in vocab:
        probabilities[word] = estimate_probability(word, previous_ngram, ngram_counts, nplus1gram_counts, vocab_size, k)
    return probabilities

# Example usage:
previous_ngram = ['i', 'eat']
vocab = closed_vocab | {'<unk>'}
probabilities = estimate_probabilities(previous_ngram, bigrams, count_ngrams(train_data_processed, 3), vocab)
print(list(probabilities.items())[:10])  # Print the probabilities for the first 10 words to check
```

# **Make Count Matrix**

```python
import numpy as np

def make_count_matrix(nplus1gram_counts, vocab):
    vocab = list(vocab)
    vocab_size = len(vocab)
    count_matrix = np.zeros((vocab_size, vocab_size))
    vocab_to_index = {word: idx for idx, word in enumerate(vocab)}

    for (ngram, count) in nplus1gram_counts.items():
        if len(ngram) == 2:  # Assuming bigrams
            prev_word, next_word = ngram
            if prev_word in vocab_to_index and next_word in vocab_to_index:
                prev_word_idx = vocab_to_index[prev_word]
                next_word_idx = vocab_to_index[next_word]
                count_matrix[prev_word_idx, next_word_idx] = count

    return count_matrix

# Example usage:
vocab = closed_vocab | {'<unk>'}
count_matrix = make_count_matrix(count_ngrams(train_data_processed, 2), vocab)
print(count_matrix)
```

# **Calculate Perplexity**

```python
def calculate_perplexity(sentence, ngram_counts, nplus1gram_counts, vocab_size, k=1):
    sentence = ['<s>'] * (len(ngram_counts) - 1) + sentence + ['<e>']
    N = len(sentence)
    log_prob_sum = 0
    for i in range(len(sentence) - len(ngram_counts) + 1):
        ngram = tuple(sentence[i:i+len(ngram_counts)])
        nplus1gram = tuple(sentence[i:i+len(ngram_counts)+1])
        log_prob_sum += np.log2(estimate_probability(nplus1gram[-1], ngram, ngram_counts, nplus1gram_counts, vocab_size, k))
    perplexity = 2**(-log_prob_sum / N)
    return perplexity

# Example usage:
sentence = ['i', 'eat', 'scrambled', 'eggs']
perplexity = calculate_perplexity(sentence, bigrams, count_ngrams(train_data_processed, 3), vocab_size)
print(f"Perplexity of the sentence '{' '.join(sentence)}': {perplexity}")
```

# **Suggest a Word**

```python
def suggest_a_word(previous_tokens, ngram_counts, nplus1gram_counts, vocab, k=1, start_with=None):
    vocab_size = len(vocab) + 1  # Include <unk>
    probabilities = estimate_probabilities(previous_tokens, ngram_counts, nplus1gram_counts, vocab, k)
    if start_with:
        filtered_probabilities = {word: prob for word, prob in probabilities.items() if word.startswith(start_with)}
    else:
        filtered_probabilities = probabilities
    suggested_word = max(filtered_probabilities, key=filtered_probabilities.get)
    return suggested_word, filtered_probabilities[suggested_word]

# Example usage:
previous_tokens = ['i', 'eat', 'scrambled']
suggested_word, max_probability = suggest_a_word(previous_tokens, bigrams, count_ngrams(train_data_processed, 3), vocab)
print(f"Suggested word: {suggested_word}, Max probability: {max_probability}")
```

# **Get Suggestions**

```python
def get_suggestions(previous_tokens, ngram_counts_list, vocab, k=1, start_with=None):
    suggestions = []
    vocab = list(vocab) + ['<unk>']

    for n in range(1, len(ngram_counts_list) + 1):
        ngram_counts = ngram_counts_list[n-1]
        previous_ngram = tuple(previous_tokens[-(n-1):])
        nplus1gram_counts = count_ngrams(train_data_processed, n+1)
        probabilities = estimate_probabilities(previous_ngram, ngram_counts, nplus1gram_counts, vocab, k)

        for word, prob in probabilities.items():
            if start_with and not word.startswith(start_with):
                continue
            suggestions.append((word, prob))

    suggestions.sort(key=lambda x: x[1], reverse=True)
    return suggestions[:5]  # Return top 5 suggestions

# Example usage:
previous_tokens = ['i', 'eat', 'scrambled']
suggestions = get_suggestions(previous_tokens, [count_ngrams(train_data_processed, n) for n in range(1, 7)], closed_vocab, k=1)
print(f"Suggestions: {suggestions}")
```
